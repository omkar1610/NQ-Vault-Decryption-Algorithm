```python
import os
import datetime
def modification_date(filename):
    t = os.path.getmtime(filename)
    return datetime.datetime.fromtimestamp(t).strftime("%Y%m%d_%H%M")

def convert(inp,out,key = 25):
    f = open(inp, "rb")
    target = open(out,"wb")
    #key = 25 #hex = 19 dec = 25
    for i in range(128):
        b = int(f.read(1).hex(),16)

        c = b^key
        # https://stackoverflow.com/questions/12638408/decorating-hex-function-to-pad-zeros    
        c = f"{c:#0{4}x}"  #padding = 4
        p = bytes.fromhex(c[2:])
        t = target.write(p)
    while True:
        b = f.read(1)
        if not b:
            # eof
            print('Done')
            break
        else:
            t = target.write(b)
    f.close()
    target.close()
```
